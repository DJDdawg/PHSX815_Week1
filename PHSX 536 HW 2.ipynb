```python
#Question 2 Linear Fit

#import packages
from matplotlib import pyplot as plt
import numpy as np
from pylab import *
from pprint import pprint
!pip install iminuit
from iminuit import Minuit
from iminuit.cost import LeastSquares

#Data from file
data = 'C:/Users/Drobe/OneDrive/Desktop/potDat.txt'
Dial_Reading, Resistance, Error = genfromtxt(data,usecols=(0,1,2),unpack=True)

#plotting data
errorbar(Dial_Reading, Resistance, yerr=Error, fmt="o", color='k', solid_capstyle='projecting', capsize=5)

#defining the fitting model function
def line(x, a, b):
    return a + b*x

#chi-square formula
least_squares = LeastSquares(Dial_Reading, Resistance, Error, line)

m = Minuit(least_squares, a=0, b=0)
m.migrad() # finds minimum of least_squares function
m.hesse() # computes errors

#plot the results
plt.errorbar(Dial_Reading, Resistance, yerr=Error, fmt="o", label="data")

plt.plot(Dial_Reading, line(Dial_Reading, *m.values), label="fit")

fit_info = [f"$\\chi^2$ / $n_\\mathrm{{dof}}$ = {m.fval:.1f} / {len(Dial_Reading) - m.nfit}",]

#Title and Axis Labels
plt.title("Linear Fit", fontsize=15, color="black")
plt.ylabel('Resistance', fontsize=15, color="black")
plt.xlabel('Potentiameter Dial Reading', fontsize=15, color="black")

for p, v, e in zip(m.parameters, m.values, m.errors):
    fit_info.append(f"{p} = ${v:.3f} \\pm {e:.3f}$")
    plt.legend(title="\n".join(fit_info));
```

```python
#Question 2 Exponential Fit

#import packages
from matplotlib import pyplot as plt
import numpy as np
from pylab import *
from pprint import pprint
!pip install iminuit
from iminuit import Minuit
from iminuit.cost import LeastSquares

#Data from file
data = 'C:/Users/Drobe/OneDrive/Desktop/potDat.txt'
Dial_Reading, Resistance, Error  = genfromtxt(data,usecols=(0,1,2),unpack=True)

#plotting data
errorbar(Dial_Reading, Resistance, yerr=Error, fmt="o", color='k', solid_capstyle='projecting', capsize=5)

#defining the fitting model function
def exponential(x, a, b, c):
    return a + b*np.exp(c*x)

#chi-square formula
least_squares = LeastSquares(Dial_Reading, Resistance, Error, exponential)

m = Minuit(least_squares, a=0, b=0, c=0)
m.migrad() # finds minimum of least_squares function
m.hesse() # computes errors

#plot the results
plt.errorbar(Dial_Reading, Resistance, yerr=Error, fmt="o", label="data")

plt.plot(Dial_Reading, exponential(Dial_Reading, *m.values), label="fit")

fit_info = [f"$\\chi^2$ / $n_\\mathrm{{dof}}$ = {m.fval:.1f} / {len(Dial_Reading) - m.nfit}",]

#Title and Axis Labels
plt.title("Exponential Fit", fontsize=15, color="black")
plt.ylabel('Resistance', fontsize=15, color="black")
plt.xlabel('Potentiameter Dial Reading', fontsize=15, color="black")

for p, v, e in zip(m.parameters, m.values, m.errors):
    fit_info.append(f"{p} = ${v:.3f} \\pm {e:.3f}$")
    plt.legend(title="\n".join(fit_info));
```

```python
#Question 2 Quadratic Fit

#import packages
from matplotlib import pyplot as plt
import numpy as np
from pylab import *
from pprint import pprint
!pip install iminuit
from iminuit import Minuit
from iminuit.cost import LeastSquares

#Data from file
data = 'C:/Users/Drobe/OneDrive/Desktop/potDat.txt'
Dial_Reading, Resistance, Error  = genfromtxt(data,usecols=(0,1,2),unpack=True)

#plotting data
errorbar(Dial_Reading, Resistance, yerr=Error, fmt="o", color='k', solid_capstyle='projecting', capsize=5)

#defining the fitting model function
def quadratic(x, a, b, c):
    return a + b*x + c*x**2

#chi-square formula
least_squares = LeastSquares(Dial_Reading, Resistance, Error, quadratic)

m = Minuit(least_squares, a=0, b=0, c=0)
m.migrad() # finds minimum of least_squares function
m.hesse() # computes errors

#plot the results
plt.errorbar(Dial_Reading, Resistance, yerr=Error, fmt="o", label="data")

plt.plot(Dial_Reading, quadratic(Dial_Reading, *m.values), label="fit")

fit_info = [f"$\\chi^2$ / $n_\\mathrm{{dof}}$ = {m.fval:.1f} / {len(Dial_Reading) - m.nfit}",]

#Title and Axis Labels
plt.title("Quadratic Fit", fontsize=15, color="black")
plt.ylabel('Resistance', fontsize=15, color="black")
plt.xlabel('Potentiameter Dial Reading', fontsize=15, color="black")

for p, v, e in zip(m.parameters, m.values, m.errors):
    fit_info.append(f"{p} = ${v:.3f} \\pm {e:.3f}$")
    plt.legend(title="\n".join(fit_info));
```

The best fit is quadratic as it has a reduced $X^2$ close to one.

Assuming that all of the error is statistical, if the fit were linear then we would expect to see a constant increase in resistance with constant increases in the Potentimeter's reading. This behavior is not reflected in the data.
